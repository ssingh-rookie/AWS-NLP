```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
tweet_data = json.loads(open('tweet_data.json').read())
```

##### return json of 2 root nodes - statuses and search_metadata

```python
tweet_data['statuses'][1]
#tweet_data['search_metadata']

#fields to be extracted below
#'created_at', 'id', 'text', 'user'/'id', 'user'/'followers_count', 'user'/'friends_count', 'retweet_count'
```

```python
tweet_list = []
for tweet in tweet_data['statuses']:
    created_at = tweet['created_at']
    tweet_id = tweet['id']
    tweet_text = tweet['text']
    user_id = tweet['user']['id']
    user_name = tweet['user']['name']
    user_followers_count = tweet['user']['followers_count']
    user_friends_count = tweet['user']['friends_count']
    retweet_count = tweet['retweet_count']
    tweet_list.append([created_at,  
                      tweet_id, 
                      tweet_text, 
                      user_id, user_name, user_followers_count, user_friends_count, 
                      retweet_count])
    
tweet_df = pd.DataFrame(tweet_list, columns =['created_at', 
                                            'tweet_id', 
                                            'tweet_text', 
                                             'user_id', 'user_name','user_followers_count','user_friends_count', 
                                             'retweet_count'])
```

```python
sentiment_response = comprehend.detect_sentiment(Text=text, LanguageCode='en')
```

```python
sentiment_response['Sentiment']
```

```python
sentiment_score = sentiment_response['SentimentScore']
```

```python
def get_sorted_sentiment(sentiment_score):
    sorted_score = sorted(sentiment_score.items(), key=lambda x: x[1], reverse=True)
    return sorted_score

def get_dominant_sentiment(sentiment_score):
    sorted_score = sorted(sentiment_score.items(), key=lambda x: x[1], reverse=True)
    return sorted_score[0]
```

#### Extract Batch sentiments (cap it to 25 requests)

```python
tweet_list = list(tweet_df['tweet_text'])[:25]
#len(tweet_list)
```

```python
batch_sa_request = {
    "LanguageCode" : "en",
    "TextList" : tweet_list
}
```

#### The code below makes batch detect sentimemt API call

```python
#batch_sa_request
batch_sa_response = comprehend.batch_detect_sentiment(TextList = tweet_list, LanguageCode='en')
```

```python
batch_sa_response
```

```python
#truncating main df to 25 rows to match the sentiment analysis batch call.
tweet_df_limted = tweet_df.iloc[:25]
```

```python
batch_sa_response['ResultList'][0]['Sentiment']
```

```python
sentiment_analysis = []
for sa_response in batch_sa_response['ResultList']:
    
    sentiment_analysis.append([
                                   sa_response.get('Index'), 
                                   sa_response.get('Sentiment'), 
                                   sa_response.get('SentimentScore')['Positive'],
                                   sa_response.get('SentimentScore')['Negative'],
                                   sa_response.get('SentimentScore')['Neutral'],
                                   sa_response.get('SentimentScore')['Mixed']
                                ]
                             )


```

```python
sentiment_analysis
```

#### Join the tweets and overall sentiment.  

```python
sa_df = pd.DataFrame(sentiment_analysis, columns=['Index', 'Sentiment', 'Positive', 'Negative', 'Neutral', 'Mixed'])
```

```python
tweet_sa_df = tweet_df_limted.join(sa_df)
```

```python
tweet_sa_df
```

```python
tweet_sa_df.to_csv('tweet_Sentiment.csv')
```
